```python
results = pd.read_sql('SELECT * FROM tbl1', con = conn).drop_duplicates()
results_out = results.sort_values("method")
results_out = results.groupby(['method']).agg({'ntest': ['mean'], 
                                                        'tr': ['mean'],
                                                        'fr': ['mean'],
                                                        'nskip': ['mean'], 
                                                        'nsample': ['mean'], 
                                                        'nsample_tot' : ['mean'], 
                                                        'true_alt_tested': ['mean'], 
                                                        'true_alt_total': ['mean'], 
                                                        'nsample_rej':['mean']})
results_out.insert(3, 'mfdr', 
                   results_out[['fr']].to_numpy() / (results_out[['fr']].to_numpy()+results_out[['tr']].to_numpy() + 0.95))
power_num = results_out['tr'].to_numpy().astype(float)
power_denom = [np.max([results_out['true_alt_tested'].to_numpy()[i].astype(float), 
                       0.0001]) for i in range(results_out.shape[0])]
results_out['power'] = power_num / np.array(power_denom)
results_out
```
